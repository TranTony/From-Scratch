# TF-IDF from scratch
## Data test: from CNN News

```python
## TF-IDF testing
docA = "The gruesome video of George Floyd's killing at the hands of a Minneapolis police officer has prompted the kind of soul searching about the role of police in society and the systemic racism in the criminal justice system that many advocates have been pushing for decades."
docB = "A rare bipartisan consensus has emerged that police reform will be necessary to respond to the groundswell."
```

```python
bowA = docA.split(" ")
bowB = docB.split(" ")
```

```python
#Create dictionary
word_dict = set(bowA).union(set(bowB))
```

```python
wordDictA = dict.fromkeys(word_dict, 0)
wordDictB = dict.fromkeys(word_dict, 0)
```

```python
#count the word in bags
for word in bowA:
    wordDictA[word]+=1

for word in bowB:
    wordDictB[word]+=1
```

```python
#calculate TF
def compute_TF(word_dict, bow):
    tf_dict = {}
    bow_count = len(bow)
    for word, count in word_dict.items():
        tf_dict[word] = count/float(bow_count)
    
    return tf_dict    

def compute_iDF(doc_list):
    import math
    idf_dict = {}
    N = len(doc_list)
    
    idf_dict = dict.fromkeys(doc_list[0].keys(), 0)
    for doc in doc_list:
        for word, count in doc.items():
            if count > 0:
                idf_dict[word]+=1

    for word, count in idf_dict.items():
        idf_dict[word] = math.log(N/float(count))
        
    return idf_dict
        
def compute_TF_iDF(bows, doc_list):
    tfidf_dict = {}
    list_tf = []
    for i, bow in enumerate(bows):
        tf_dict = compute_TF(doc_list[i], bow)
        list_tf.append(tf_dict)

    idf_dict = compute_iDF(doc_list)
    tfidf_dict_list = []
    
    for tf_bow in list_tf:
        tfidf_dict = {}
        for word, val in tf_bow.items():
            tfidf_dict[word] = val*idf_dict[word]
        tfidf_dict_list.append(tfidf_dict)
    return tfidf_dict_list
    
tfidf_dict_list = compute_TF_iDF([bowA, bowB], [wordDictA, wordDictB])
    
```
